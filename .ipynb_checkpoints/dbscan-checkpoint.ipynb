```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
import joblib

# Load Data
hot_100 = pd.read_csv("extended_hot_100.csv")
not_hot_100 = pd.read_csv("extended_not_hot.csv")

# Concat
all_songs = pd.concat([hot_100, not_hot_100], ignore_index=True)

# Data Cleaning
all_songs = all_songs.dropna(subset=['valence', 'danceability'])

# Scaling Features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(all_songs[['valence', 'danceability']])

# KMeans Clustering
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
all_songs['KMeans_Cluster'] = kmeans.fit_predict(scaled_features)

# Visualizing KMeans Clusters
plt.scatter(scaled_features[:, 0], scaled_features[:, 1], c=all_songs['KMeans_Cluster'], cmap='viridis')
plt.xlabel('Valence')
plt.ylabel('Danceability')
plt.title('KMeans Clustering of Songs')
plt.show()

# DBSCAN Clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
all_songs['DBSCAN_Cluster'] = dbscan.fit_predict(scaled_features)

# Visualizing DBSCAN Clusters
plt.scatter(scaled_features[:, 0], scaled_features[:, 1], c=all_songs['DBSCAN_Cluster'], cmap='plasma')
plt.xlabel('Valence')
plt.ylabel('Danceability')
plt.title('DBSCAN Clustering of Songs')
plt.show()

# Save Models and Data
joblib.dump(kmeans, 'kmeans_model.pkl')
joblib.dump(dbscan, 'dbscan_model.pkl')
all_songs.to_csv('all_songs_with_clusters.csv', index=False)
```

```python
# Scaling Features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(all_songs[['valence', 'danceability']])

# Determine the optimal number of clusters using the Elbow Method
wcss = []  
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

# Plotting the Elbow Method Graph
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Optimal number of clusters
optimal_k = 3

# KMeans Clustering with optimal_k
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', max_iter=300, n_init=10, random_state=0)
all_songs['KMeans_Cluster'] = kmeans.fit_predict(scaled_features)
```

# Analysis:

### KMeans Clustering:
* The KMeans algorithm divided the data into 3 distinct clusters.
* KMeans tries to form clusters of equal sizes and assumes spherical shapes of clusters.
* Given that it has created distinct clusters, it seems suitable for identifying groupings within the data based on 'valence' and 'danceability'.

### DBSCAN Clustering:
*  DBSCAN identified all data points as a single cluster.



# Conclusion:
* KMeans: the clusters are well-separated and cohesive within themselves .
* KMeans has found meaningful groupings of songs based on emotion (valence) and rhythm (danceability).
* DBSCANS: the single cluster outcome suggests that DBSCAN is not suitable for this particular dataset, due to uniform data density.
